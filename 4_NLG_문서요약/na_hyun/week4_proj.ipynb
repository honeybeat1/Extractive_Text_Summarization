```python
import re

def preprocess(string):
    str_list = string.split('\n\n') # \n\n 기준으로 분리
    str_list[0] = str_list[0].replace('[스포탈코리아] ', '') # [스포탈코리아] 삭제
    
    result = []
    for strg in str_list:
        strg = re.sub(r'<[^>]*>', '', strg) # html 제거
        strg = re.sub(r'\([^)]*\)|\[[^)]*\]|\<[^)]*\>', ' ', strg) # 괄호와 그 안에 문자 제거

        if '기자' in strg: # 기자 이름이 들어간 경우 기자이름 삭제
            strg = strg.split('기자')[-1]
        if '다.' in strg: # 문장 단위로 분리
            strg = strg.split('. ')
            for s in strg:
                if s != '':
                    result.append(s)
            
    return ' '.join(result)
```

```python
def special(string):
    regex = r"[^가-힣a-zA-Z0-9 ]" # 특수 문자 제거
    string = re.sub(regex, '', string)

    return string
```

```python
import torch
from transformers import PreTrainedTokenizerFast
from transformers import BartForConditionalGeneration
```

```python
# device type
if torch.cuda.is_available():
    device = torch.device("cuda")
    print(f"# available GPUs : {torch.cuda.device_count()}")
    print(f"GPU name : {torch.cuda.get_device_name()}")
else:
    device = torch.device("cpu")
print(device)
```

```python
tokenizer = PreTrainedTokenizerFast.from_pretrained('digit82/kobart-summarization')
model = BartForConditionalGeneration.from_pretrained('digit82/kobart-summarization').to(device)
```

```python
df = df.reset_index()
df = df[['TITLE', 'CONTENT', 'PUBLISH_DT', 'PRE_CONTENT', 'len']]
```

```python
df
```

```python
text = df['PRE_CONTENT'][500]
text
```

```python
raw_input_ids = tokenizer.encode(text)
input_ids = [tokenizer.bos_token_id] + raw_input_ids + [tokenizer.eos_token_id]

summary_ids = model.generate(torch.tensor([input_ids]).to(device),  num_beams=4,  max_length=512,  eos_token_id=1)
tokenizer.decode(summary_ids.squeeze().tolist(), skip_special_tokens=True)
```

```python
text_list = df['PRE_CONTENT'].values
df['KoBART'] = [0] * len(text_list)

for i, text in enumerate(text_list):
    if df['len'].iloc[i] < 2000:
        try:
            raw_input_ids = tokenizer.encode(text)
            input_ids = [tokenizer.bos_token_id] + raw_input_ids + [tokenizer.eos_token_id]

            summary_ids = model.generate(torch.tensor([input_ids]).to(device),  num_beams=4,  max_length=512,  eos_token_id=1)
            df['KoBART'].iloc[i] = tokenizer.decode(summary_ids.squeeze().tolist(), skip_special_tokens=True)
        except :
            continue
```

```python
df.to_csv('KoBART_최종.csv')
```
